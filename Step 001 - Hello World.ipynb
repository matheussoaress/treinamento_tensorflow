## Full Connected Neural Network / Dense Neural Network / Standard Neural Network

Uma rede neural densa, ou totalmente conectada, é a rede neural artifical em que todas as empresas são conectadas à todos os neurônios da primeira camada, que por sua vez são conectados a todos os neurônios das próximas camadas ou a camada de saída.

### Base de dados

```python
# ax + b = y

X = np.array([-1, 0, 1, 2, 3, 4])
y = np.array([-3, -1, 1, 3, 5, 7])
```

```python
# Limpando sessão e setando semente

tf.keras.backend.clear_session()
tf.random.set_seed(51)
np.random.seed(51)
```

```python
model = tf.keras.Sequential()
```

```python
# Adicionando 1 camada com 1 neurônio

model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```python
#Compilando o modelo

model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
### Entendendo o modelo

model.summary()
```

```python
history = model.fit(X, y, epochs=500)
```

```python
import matplotlib.pyplot as plt

def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.show()
```

```python
plot_graphs(history, 'loss')
```

```python
# Predizendo um valor

model.predict([[10]])
```
